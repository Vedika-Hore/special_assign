```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))

plt.subplot(2,2,1)
sns.histplot(df["lead_time"],color="skyblue",edgecolor="black")
plt.title("Histogram of lead_time")
plt.xlabel("lead_time")
plt.ylabel("count")

plt.subplot(2,2,2)
sns.histplot(df["arrival_date_year"],color="skyblue",edgecolor="black")
plt.title("Histogram of arrival_date_year")
plt.xlabel("arrival_date_year")
plt.ylabel("count")

plt.subplot(2,2,3)
sns.histplot(df["agent"],color="skyblue",edgecolor="black")
plt.title("Histogram of agent")
plt.xlabel("agents")
plt.ylabel("count")

plt.subplot(2,2,4)
sns.histplot(df["company"],color="skyblue",edgecolor="black")
plt.title("Histogram of company")
plt.xlabel("company")
plt.ylabel("count")

plt.tight_layout()

```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
sns.countplot(x=df["customer_type"],color="pink",edgecolor="black")
plt.title("Countplot of sentence length")

plt.subplot(1,3,2)
sns.countplot(x=df["reservation_status"],color="lightgreen",edgecolor="black")
plt.title("Countplot of label")

plt.subplot(1,3,3)
sns.countplot(x=df["deposit_type"],color="skyblue",edgecolor="black")
plt.title("Countplot of source")

plt.tight_layout()
```

```python
corr=df.select_dtypes(include=('number')).corr()
sns.heatmap(corr,annot=True,cmap="viridis")
plt.title("Heatmap correlation")
```

```python
sns.scatterplot(x=df["arrival_date_year"],y=df["company"],edgecolor="black")
plt.title("Scatterplot")
```

```python
sns.boxplot(
    x="customer_type",
    y="agent",data=df,
    legend=False)
```

```python
counts=df["customer_type"].value_counts()
plt.figure(figsize=(5,5))
plt.pie(
    counts,
    labels=counts.index,
    autopct="%1.1f%%",
    #colors=["skyblue", "lightgreen"],
    startangle=90
)
plt.title("Distribution of Cutomer Type")
plt.show()
plt.tight_layout()
```
