```python
filepath=("/content/car_price_dataset.csv")
df=pd.read_csv(filepath)
print(df)
```

```python
print("Independent Feature:")
X=df.drop("Price",axis=1)
X
```

```python
print("Dependent Feature:")
Y=df["Price"]
Y
```

```python
print("Shape of X=",X.shape)
```

```python
print("Shape of Y=",Y.shape)
```

```python
X=df.select_dtypes(include=["int64"])
encoded_df=pd.get_dummies(data=df,dtype=int)
encoded_df.astype(int)
encoded_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
num_cols=["Year","Engine_Size","Mileage","Doors","Price"]
scaler=StandardScaler()
encoded_df[num_cols]=scaler.fit_transform(encoded_df[num_cols])
encoded_df[num_cols]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(encoded_df,Y,test_size=0.33,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
print("intercept=",model.intercept_)
print("Coefficient=",model.coef_)
```

```python
Y_pred = model.predict(X_test)
print("Predicted value:",Y_pred[0:11])
print("Actual value:",Y_test.tolist()[0:11])
```

```python
from sklearn.metrics import r2_score
r2=r2_score(Y_test,Y_pred)
print("R Square=",r2)
```

```python
import joblib
joblib.dump(model,"model.pkl")
joblib.dump(scaler,"scaler.pkl")
joblib.dump(X_train.columns.to_list(),"column.pkl")
```
