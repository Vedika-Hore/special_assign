```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
plt.subplot(2,2,1)
sns.histplot(df["age"],edgecolor="black")
plt.title("Histogram of Age")
plt.xlabel("age")
plt.ylabel("count")


plt.subplot(2,2,2)
sns.histplot(df["bmi"],edgecolor="black")
plt.title("Histogram of BMI")
plt.xlabel("BMI")
plt.ylabel("count")

plt.subplot(2,2,3)
sns.histplot(df["children"],edgecolor="black")
plt.title("Histogram of children")
plt.xlabel("children")
plt.ylabel("count")

plt.subplot(2,2,4)
sns.histplot(df["expenses"],edgecolor="black")
plt.title("Histogram of Expenses")
plt.xlabel("expenses")
plt.ylabel("count")

plt.tight_layout()
```

```python
# Age:
# The ages are spread across different age groups, with most customers
# between 20 and 60 years. The distribution is fairly uniform.

# BMI:
# Most BMI values are between 25 and 35, indicating that many customers
# are overweight. Very few customers have extremely low or high BMI values.

# Children:
# Most customers have 0 to 2 children. The number of customers decreases
# as the number of children increases.

# Expenses:
# Insurance expenses are right-skewed. Most customers have lower expenses,
# while a small number have very high medical charges.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))

plt.subplot(1,3,1)
sns.countplot(x="sex", data=df)
plt.title("Sex Count")

plt.subplot(1,3,2)
sns.countplot(x="smoker", data=df)
plt.title("Smoker Count")

plt.subplot(1,3,3)
sns.countplot(x="region", data=df)
plt.title("Region Count")

plt.tight_layout()

```

```python
# Smoker:
# Most customers are non-smokers, while smokers make up a smaller group.
# This indicates that the majority of insurance customers do not smoke.

# Sex:
# The number of male and female customers is almost equal.
# This shows that the dataset is well balanced by gender.

# Region:
# Customers are distributed across all four regions.
# The Southeast region has the highest number of customers, while the other
# regions have similar customer counts.

# Overall Market Trends:
# - Most insurance customers are non-smokers.
# - Male and female customers are nearly equally represented.
# - Customers come from all regions, with the Southeast having the largest share.
# - The dataset represents a diverse customer base for insurance analysis.
```

```python
corr=df.select_dtypes(include=['number']).corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
```

```python
#Observations:

# 1. The heatmap shows the correlation between all numeric features.
# 2. Darker colors indicate stronger positive or negative correlation.
# 3. Features with correlation values closer to 1 are strongly positively correlated with price.
# 4. Features with values closer to -1 are strongly negatively correlated with price.
# 5. Features with values close to 0 have little or no correlation with price.
# 6. The feature(s) with the highest correlation value with price have the strongest influence on car price.
```
