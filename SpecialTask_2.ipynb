```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(4,4))
sns.histplot(df["complexity_score"])
plt.title("Complexity Score")
plt.xlabel("Complexity_score")
plt.ylabel("Count")
```

```python
sns.barplot(x="sentence_length",y="complexity_score",data=df)
plt.title("Sample Line Chart")
```

```python
sns.scatterplot(x="sentence_length",y="complexity_score",data=df)
plt.title("Sentence length and Complexity Score")
plt.xlabel("complexity_score")
plt.ylabel("sentence_length")
```

```python
sns.boxplot(x="sentence_length",y="complexity_score",data=df)
plt.title("Sentence length and Complexity Score")
plt.xlabel("complexity_score")
plt.ylabel("sentence_length")
```

```python
sns.heatmap(df.select_dtypes(include=("number")).corr(),annot=True,cmap="coolwarm")
```

```python
df["score_group"] = pd.cut(df["complexity_score"],
                           bins=[0, 0.4, 0.6, 1],
                           labels=["Low", "Medium", "High"])

df["score_group"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Complexity Score Groups")
plt.ylabel("")
plt.show()
```
