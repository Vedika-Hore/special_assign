```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))

plt.subplot(2,2,1)
sns.histplot(df["id"],color="skyblue",edgecolor="black")
plt.title("Histogram of ID")
plt.xlabel("id")
plt.ylabel("count")

plt.subplot(2,2,2)
sns.histplot(df["complexity_score"],color="skyblue",edgecolor="black")
plt.title("Histogram of complexity score")
plt.xlabel("complexity_score")
plt.ylabel("count")

plt.subplot(2,2,3)
sns.histplot(df["grammar_errors_count"],color="skyblue",edgecolor="black")
plt.title("Histogram of grammar errors count")
plt.xlabel("igrammar_errors_count")
plt.ylabel("count")

plt.subplot(2,2,4)
sns.histplot(df["sentence_length"],color="skyblue",edgecolor="black")
plt.title("Histogram of sentence length")
plt.xlabel("sentence_length")
plt.ylabel("count")

plt.tight_layout()

```

```python

plt.subplot(1,3,1)
sns.countplot(x=df["sentence_length"],color="pink",edgecolor="black")
plt.title("Countplot of sentence length")

plt.subplot(1,3,2)
sns.countplot(x=df["label"],color="lightgreen",edgecolor="black")
plt.title("Countplot of label")

plt.subplot(1,3,3)
sns.countplot(x=df["source"],color="skyblue",edgecolor="black")
plt.title("Countplot of source")

plt.tight_layout()

```

```python
corr=df.select_dtypes(include=('number')).corr()
sns.heatmap(corr,annot=True,cmap="viridis")
plt.title("Heatmap correlation")
```

```python
sns.scatterplot(x=df["sentence_length"],y=df["id"],edgecolor="black")
plt.title("Scatterplot")

```

```python
sns.boxplot(
    x="label",
    y="complexity_score",
    hue="label",
    data=df,
    palette={"Human": "blue", "AI": "green"},
    legend=False)
```

```python
counts=df["label"].value_counts()
plt.figure(figsize=(5,5))
plt.pie(
    counts,
    labels=counts.index,
    autopct="%1.1f%%",
    colors=["skyblue", "lightgreen"],
    startangle=90
)
plt.title("Distribution of Human vs AI")
plt.show()
```
